Date column is showing object in datatypes hence it is convert in dat format

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df.dtypes
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df.groupby('Holiday_Flag')['Weekly_Sales'].mean()
```

```python
monthly_sales = df.groupby('Month')['Weekly_Sales'].sum()

monthly_sales.plot(kind='bar')
plt.title("Monthly Sales")
plt.show()
```

```python
yearly_sales = df.groupby('Year')['Weekly_Sales'].sum()

yearly_sales.plot(kind='line', marker='o')
plt.title("Yearly Sales Trend")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.histplot(df['Weekly_Sales'], bins=30)
plt.show()
```

```python
sns.boxplot(x=df['Weekly_Sales'])
plt.show()
```

```python
top_stores = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(10)

top_stores.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Stores")
plt.ylabel("Sales")
plt.show()
```

```python
top_stores = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(20)

top_stores.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Stores")
plt.ylabel("Sales")
plt.show()
```

```python
top_stores = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(50)

top_stores.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Stores")
plt.ylabel("Sales")
plt.show()
```

```python
top_stores = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(20)

top_stores.plot(kind='barh', figsize=(10,5))
plt.title("Top 10 Stores")
plt.ylabel("Sales")
plt.show()
```
